# Stablecoin Example

Results of running a simple Uniswap trading model with 10,000 agents for 500 steps.  This example uses 
zero intelligence agents randomly swapping DIA and USDC.  The results of the model were saved in the `avro` 
format and plotted using data frames.

```python
import polars as pl

df = pl.read_avro('usdc_dia_10000_500.avro')
print(df)
```

```python
df.plot.area(x="steps", y=["DIA_vol", "USDC_vol"],legend='bottom_right', title="Vol") 
```

```python
df.plot.area(x="steps", y=["DIA_reserve", "USDC_reserve"],legend='bottom_right', title="Pool Reserves") 
```

```python
df.plot.line(x="steps", y=["DIA_price", "USDC_price"], title="Price")  
```
